```python

import numpy as np
import gymnasium as gym
import panda_gym
import matplotlib.pyplot as plt
from tqdm import tqdm
import torch

from td3_algo import TD3Trainer
from sac_algo import SACTrainer
from snaf_algo import SNAFTrainer

import gymnasium as gym
import panda_gym

env = gym.make("PandaReach-v3", render_mode="rgb_array")

env.reset()
plt.axis('off')
plt.imshow(env.render())
plt.show()
print('\nObservation Space:', env.observation_space)
print('\nAction Space: ', env.action_space)
print('\nInitial State: ',env.reset())
print('\nA step in the environment: ', env.step(np.array(1.0)))
print('\n\nRendered Environment: ')
```

```python
obs_shape = env.observation_space['observation'].shape[0] + \
            env.observation_space['achieved_goal'].shape[0] + \
            env.observation_space['desired_goal'].shape[0]


expert_naf = NAFTrainer(env=env, input_dims=obs_shape, agent_name='NAF', model_save_path='./Models/',
                    exploration_period=300)
score_history_naf, avg_score_history_naf = expert_naf.naf_train(n_episodes=1000, opt_steps=10, print_every=100,
                                                    plot_save_path='../Results/NAF_Performance.png')



expert_td3 = TD3Trainer(env=env, input_dims=obs_shape, agent_name='TD3', model_save_path='./Models/',
                    exploration_period=300)
score_history_td3, avg_score_history_td3 = expert_td3.td3_train(n_episodes=1000, opt_steps=10, print_every=100,
                                                    plot_save_path='../Results/TD3_Performance.png')



expert_sac = SACTrainer(env=env, input_dims=obs_shape, agent_name='Expert', model_save_path='./Models/',
                    exploration_period=300)
score_history_sac, avg_score_history_sac = expert_sac.sac_train(n_episodes=1000, opt_steps=10, print_every=100,
                                                    plot_save_path='../Results/SAC_Performance.png')

```

```python
import matplotlib.pyplot as plt
import numpy as np


# Function to calculate moving average
def moving_average(data, window_size=10):
    if len(data) < window_size:
        return np.array([])  # Avoid calculation if data is smaller than the window size
    return np.convolve(data, np.ones(window_size) / window_size, mode='valid')

# Calculate moving averages for smooth mean rewards with a window size of 10
mean_snaf = moving_average(score_history_snaf, window_size=10)
mean_sac = moving_average(score_history_sac, window_size=10)
mean_td3 = moving_average(score_history_td3, window_size=10)

# Plotting
plt.figure(figsize=(10, 6))

# Individual rewards with less transparency
plt.plot(score_history_snaf, label="SNAF Reward", alpha=0.3, color="blue", linestyle='--')
plt.plot(score_history_sac, label="SAC Reward", alpha=0.3, color="green", linestyle='--')
plt.plot(score_history_td3, label="TD3 Reward", alpha=0.3, color="red", linestyle='--')

# Mean rewards with solid lines
plt.plot(range(len(mean_snaf)), mean_snaf, label="Mean SNAF Reward", color="blue", linewidth=2)
plt.plot(range(len(mean_sac)), mean_sac, label="Mean SAC Reward", color="green", linewidth=2)
plt.plot(range(len(mean_td3)), mean_td3, label="Mean TD3 Reward", color="red", linewidth=2)

# Customize the plot
plt.title("Rewards Comparison")
plt.xlabel("Episodes")
plt.ylabel("Reward")
plt.legend()
plt.grid(True)
plt.tight_layout()

# Show the plot
plt.show()
```
